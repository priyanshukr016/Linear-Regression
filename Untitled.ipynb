```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df["cgpa"],df["package"],color='g')
plt.xlabel("cgpa")
plt.ylabel("package")
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
x = df.drop(columns="package")
x
```

```python
y = df.drop(columns="cgpa")
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_test.iloc[0]
```

```python
lr.predict(X_test.iloc[[0]])
```

```python
plt.scatter(df["cgpa"],df["package"],color='black')
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel("cgpa")
plt.ylabel("package")
plt.show()
```

### y = mx+b

```python
m = lr.coef_
b = lr.intercept_
```

```python
x = float(input("cgpa"))
y = m*x+b
print("your package will be around: ",y)
```
